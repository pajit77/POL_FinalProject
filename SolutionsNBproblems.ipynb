```python
import pandas as pd
```

## Tennis Data

```python
df = pd.read_excel("ProblemData.xlsx")
#xl.sheet_names
```

```python
df
```

```python
def count_creator(data):
    cols = data.columns
    outcome = cols[-1]
    cols = cols[:-1]
    cols = list(reversed(cols))
    counter = 0

    for col in cols:

        categories = list(data[col].unique())
        data2 = data.groupby([col,outcome]).size().reset_index(name='counts')
        data2 = data2.pivot(index=col, columns=outcome, values='counts')
        data2.reset_index(inplace=True)
        data2.columns = ['Column','no','yes']
        data2.fillna(0,inplace=True)
        if counter == 0:
            counter+=1
            Counts_Table = data2
        else:
            Counts_Table = data2.append(Counts_Table)

    total_yes = int(data2['yes'].sum())
    total_no = int(data2['no'].sum())
    
    Counts_Table.no = Counts_Table.no.astype(int)
    Counts_Table.yes = Counts_Table.yes.astype(int)

    prior_yes =total_yes/(total_yes+total_no)
    prior_no = total_no/(total_no+total_yes)

    Counts_Table['cond_yes'] = Counts_Table['yes'].astype(str) + str('/') + str(total_yes) 
    Counts_Table['cond_no'] = Counts_Table['no'].astype(str) + str('/') + str(total_no) 
    
    ordercols = ['Column','yes','no','cond_yes','cond_no']
    Counts_Table = Counts_Table[ordercols]
    Counts_Table.fillna(0,inplace=True)
    return Counts_Table,prior_yes,prior_no,total_yes,total_no
```

## Counts and Conditional Probabilities for Play Tennis

```python
Counts_Table,prior_yes,prior_no,total_yes,total_no = count_creator(df)
Counts_Table
```

```python
print ("Yes Count = ", total_yes)
print ("No Count = ",total_no)

print ("Prior of Yes = " ,prior_yes )
print ("Prior of No = " ,prior_no )
```

```python
L_nogiven_overcast_hot_normal_true = 0 * 0.4 * 0.2 * 0.6 * prior_no
L_yesgiven_overcast_hot_normal_true = 0.444444 * 0.222222 * 0.666667 * 0.333333 * prior_yes

print("\n")

print ("Likelihood of No = " ,L_nogiven_overcast_hot_normal_true )
print ("Likelihood of Yes = " ,L_yesgiven_overcast_hot_normal_true )
```
